```python
def add(x, y):
    """ add function"""
    return x + y

def subtract(x, y):
    """ add function"""
    return x - y

def multiply(x, y):
    """ add function"""
    return x * y

def divide(x, y):
    """ add function"""
    if y == 0:
        raise ValueError('cannot be divided by zero')
    return x / y
```
