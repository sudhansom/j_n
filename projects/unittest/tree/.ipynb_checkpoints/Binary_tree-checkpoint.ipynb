```python
class Bst:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
    def add_child(self, data):
        if self.data == data:
            return
        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = Bst(data)
        if data > self.data:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = Bst(data)
    def in_order_traversal(self):
            elements = []
            if self.left:
                elements += self.left.in_order_traversal()
            elements.append(self.data)
            if self.right:
                elements += self.right.in_order_traversal()
            return elements
    def depth_first_search(self):
        print(self.data)
        if self.left:
            self.left.depth_first_search()
        if self.right:
            self.right.depth_first_search()
    def find_closest_value(self,  target, closest='', diff=100):
        if self.data == target:
            return self.data
        if abs(self.data - target) < diff:
            diff = abs(self.data - target)
            closest = self.data
        if target < self.data:
            if self.left:
                closest = self.left.find_closest_value(target, closest, diff)
        if target > self.data:
            if self.right:
                closest = self.right.find_closest_value(target, closest, diff)
        return closest
    def branch_sum(self, sums=0, arr=[]):
        sums += self.data
        if self.left is None and self.right is None:
            arr.append(sums)
        if self.left:
            self.left.branch_sum(sums, arr)
        if self.right:
            self.right.branch_sum(sums, arr)
        return arr
    def max_node_depth(self, cnt=-1, arr=[]):
        cnt += 1
        if self.left is None and self.right is None:
            arr.append(cnt)
        if self.left:
            self.left.max_node_depth(cnt, arr)
        if self.right:
            self.right.max_node_depth(cnt, arr)
        return max(arr)
    def sum_of_node_depth(self, cnt=-1, arr=[]):
        cnt += 1
        arr.append(cnt)
        if self.left:
            self.left.sum_of_node_depth(cnt, arr)
        if self.right:
            self.right.sum_of_node_depth(cnt, arr)
        return sum(arr)
    def sum_of_node(self, cnt=-1, sums=0):
        cnt += 1
        sums += cnt
        if self.left:
            sums = self.left.sum_of_node(cnt, sums)
        if self.right:
            sums = self.right.sum_of_node(cnt, sums)
        return sums
    def no_of_nodes(self, total=0):
        total += 1
        if self.left:
            total = self.left.no_of_nodes(total)
        if self.right:
            total = self.right.no_of_nodes(total)
        return total
        
        
        
    
```

```python
def build_a_tree(arr):
    root = Bst(arr[0])
    for items in arr[1:]:
        root.add_child(items)
    return root

arr = [17, 4, 9, 20, 1, 18, 23, 34]
root = build_a_tree(arr)
print("in order traversal :")
print(root.in_order_traversal())
print("depth first search: ")
root.depth_first_search()
print(f"closest value : {root.find_closest_value(21)}")
print(f"Branch sum : {root.branch_sum()}")
print(f"maximum node depth : {root.max_node_depth()}")
print(f"Sum of all nodes depth:{root.sum_of_node_depth()}")
print(f"sum of all depths: {root.sum_of_node()}")
print(f"total number of nodes: {root.no_of_nodes()}")
```
